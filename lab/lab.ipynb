```python
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of X_train[i]:", X_train[0].shape)  # Shape of a single image
print("Shape of X_test[i]:", X_test[0].shape)    # Shape of a single image
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Reshape data into Nx784 format
X_train_flat = X_train.reshape(X_train.shape[0], 784)
X_test_flat = X_test.reshape(X_test.shape[0], 784)

# Normalize pixel values (gives better performance?)
X_train_flat = X_train_flat / 255.0
X_test_flat = X_test_flat / 255.0
```

```python
# Try different values of k
k_values = [1, 3, 5, 7, 9]
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_flat, y_train)
    y_pred = knn.predict(X_test_flat)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"k={k}, Accuracy={acc:.4f}")

# Find the best k
optimal_k = k_values[np.argmax(accuracies)]
print(f"Optimal k: {optimal_k}")
```

```python
# Train kNN with optimal k
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)
knn_optimal.fit(X_train_flat, y_train)
y_pred_optimal = knn_optimal.predict(X_test_flat)

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred_optimal)
print(f"Final accuracy with k={optimal_k}: {accuracy:.4f}")

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_optimal)

# Plot confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix for kNN Classifier")
plt.show()
```

The confusion matrix reveals the most common misclassifications - 4 and 9, 7 and 1, and 3 and 8 are confused the most, likely due to their visual similarities. As for patterns, it seems like 0 is the least misclassified number, while 8 is the most.
